## For catgorical columns using countplot 

```python
print('total male and female values are \n',df['sex'].value_counts(),end='\n'*3)
plt.figure(figsize=(7,7))
sns.countplot(df.sex)
plt.show()
```

```python
print('Total values of different island in data are \n',df['island'].value_counts(),end='\n'*3)
plt.figure(figsize=(7,7))
sns.countplot(df.island)
plt.show()
```

```python
print('Total values of different species of penguine in data are \n',df['species'].value_counts(), end='\n'*3)
plt.figure(figsize=(7,7))
sns.countplot(df.species)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.swarmplot(x='species',y='body_mass_g',data=df)
plt.show()
```

## Conclusion 2 - From above plot Mostly Gentoo species has higher mass than others

```python
plt.figure(figsize=(10,7))
sns.swarmplot(x='species',y='body_mass_g',hue='sex',data=df)
plt.show()
```

##  Conclusion 3 - From above plot For all species mostly male are heavier than female.

```python
plt.figure(figsize=(10,7))
sns.swarmplot(x='species',y='flipper_length_mm',data=df)
```
